```python
d1=Data.head(16384)
```

```python
df = d1["amplitude-XP1"]
N= len(df)
y=[]
for j in range(0,(int(N/2)+1)):
    R=0    
    for i in range(0,(int(N/2))):
        R = R+ df[i]*df[i+j]
    y.append(R*2/N)
    
# Autocorrelation coefficient calculation

df = pd.DataFrame(y)
df.columns =['R']
R0=df.iloc[0,0]
R0
df['A'] = df['R']/R0
dt = d1["time"]
t=[]
for i in range(len(df)):
    t.append(dt[i])
    
df['t']=t
# Plot waveform signal
fig = plt.figure(figsize=(8,8), dpi=72)
x=d1["time"]
y=d1["amplitude-XP1"]
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("amplitude-XP1")
plt.title("Original Signal waveform")
fig.savefig("Signal wave form.png")
plt.grid()
# Plot figure of Autocorrelation coefficient
fig1 = plt.figure(figsize=(8,8), dpi=72)
x1=df["t"]
y1=df["A"]
plt.plot(x1,y1)
plt.xlabel("Time[ms]")
plt.ylabel("Amplitude[g]")
plt.title("Autocorrelation coefficient")
fig1.savefig("Autocorrelation coefficient.png")
plt.grid()
```

```python
from numpy.fft import fft, fftfreq

N=df.size # Number of samples

Duration =0,12 # Length of the wavwform in second(Duration)
Te = 1/16384 # Sampling time

X = fft(df)  # Fast Fourier Transform

freq = fftfreq(df.size, 1/16384)  # Fast Fourier Transform Frequency

# Take only positif absolute value and  normalisation

X_abs = np.abs(X[:N//2])*2.0/N

freq_pos = freq[:N//2]

peakX = np.max(X_abs) # Find max peak

locX = np.argmax(X_abs) # Find its location

freq_posX = freq_pos[locX] # Get the actual frequency value

# New - Plot the max point
#plt.plot(freq_posX, peakX, 'r.', markersize=9)


fig = plt.figure(figsize=(10,7))

plt.plot(freq_posX, peakX, 'r.', markersize=9)

freq_pos = freq[:N//2]
plt.plot(freq_pos, X_abs, label="Amplitude absolue")
plt.xlim(1, 450)  # zoom frequency
plt.ylim(0, 0.15)
plt.grid()
plt.xlabel("FrÃ©quence (Hz)")
plt.ylabel("Amplitude |X(f)| [mm/s]_RMS")

plt.show()
fig.savefig("Spectre FFT")
```
